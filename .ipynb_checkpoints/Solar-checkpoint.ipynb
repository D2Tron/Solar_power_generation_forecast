```python
#To split the data into training and testing
import pandas as pd

data = pd.read_csv("solar.csv")

trainData = data[data['TIMESTAMP'] <= '20130701 00:00']
testData = data[data['TIMESTAMP'] > '20130701 00:00']

trainData.to_csv('solar_training.csv', index=False)
testData.to_csv('solar_test.csv', index=False)
```

```python
#For encoding to be used by classifiers instead of Regressors
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from sklearn import utils

trainData = pd.read_csv("solar_training.csv")
testData = pd.read_csv("solar_test.csv")

X_train1 = trainData[trainData['ZONEID'] == 1]
y_train1 = X_train1['POWER']
X_train1 = X_train1.drop(X_train1.columns[[0, 1, 14]], axis=1)

X_test1 = testData[testData['ZONEID'] == 1]
y_test1 = X_test1['POWER']
X_test1 = X_test1.drop(X_test1.columns[[0, 1, 14]], axis=1)

lab_enc = preprocessing.LabelEncoder()
y_train1E = lab_enc.fit_transform(y_train1)
y_test1E = lab_enc.fit_transform(y_test1)
```

```python
#PCA if necessary
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd

X_train1S = StandardScaler().fit_transform(X_train1)
X_test1S = StandardScaler().fit_transform(X_test1)

pca = PCA(n_components=4)
pCompTrain1 = pca.fit_transform(X_train1S)
pCompTest1 = pca.fit_transform(X_test1S)
X_train1P = pd.DataFrame(data = pCompTrain1, columns = ['p1', 'p2', 'p3', 'p4'])
X_test1P = pd.DataFrame(data = pCompTest1, columns = ['p1', 'p2', 'p3', 'p4'])
```

```python
#Regression model for plant no. 1
from sklearn.ensemble import RandomForestRegressor
import math

X_train1 = trainData[trainData['ZONEID'] == 1]
y_train1 = X_train1['POWER']
X_train1 = X_train1.drop(X_train1.columns[[0, 1, 14]], axis=1)

X_test1 = testData[testData['ZONEID'] == 1]
y_test1 = X_test1['POWER']
X_test1 = X_test1.drop(X_test1.columns[[0, 1, 14]], axis=1)

regr = RandomForestRegressor(max_depth=10, random_state=0)
regr.fit(X_train1, y_train1)

predicted = regr.predict(X_test1)

numSamples = len(y_test1)

y_test1.reset_index(drop=True, inplace=True)

diff = 0
root = 0
for i in range(numSamples):
    diff += abs(y_test1[i] - predicted[i])
    root += pow(abs(y_test1[i] - predicted[i]), 2)

diff = diff/numSamples
root = root/numSamples
print(diff)
print(math.sqrt(root))
```
